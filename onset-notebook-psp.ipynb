# PSP/ISOIS EPIHI (~1 MeV electrons and >25 MeV protons)

- ``SunpyUserWarning`` that a variable has been skipped because it has more than 2 dimensions can be ignored

```python
event_onset_epihi_e = Event(spacecraft='psp', sensor='isois-epihi', data_level='l2', species='e', start_date=datetime.date(2021,10,28), end_date=datetime.date(2021,10,30), data_path=path)
```
