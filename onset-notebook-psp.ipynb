# PSP/ISOIS EPILO (~100 keV electrons)

- only electrons so far
- For EPILO, the energy tables have been changed on June 14, 2021! That is, the energy values for each channel are different before and afterwards. Maybe the energy channel selection has to be adjusted for earlier events accordingly. EPIHI (i.e. our 1 MeV electrons and >25 MeV protons) is not affected.
- ``viewing`` directions are ``'0'`` to ``'7'``, with ``'3'`` usually pointing sunward and ``'7'`` anti-sunward. The other directions are rather poorly defined.
- ``threshold`` functionality like for Wind/3DP, see example below. Be very careful with setting a ``threshold`` because it's applied to **all** viewing directions and energies!

Note: In the following a different event is used compared to the one for EPIHI above.

## ``threshold=None`` (standard)

```python
event_onset_epilo = Event(spacecraft='psp', sensor='isois-epilo', data_level='l2', species='e', start_date=datetime.date(2021,10,28), end_date=datetime.date(2021,10,30), data_path=path, threshold=None)
```

Get meta data like energies with ``event_onset_epilo.current_e_energies``

```python
flux_series_epilo, onset_stats_epilo, onset_found_epilo, peak_flux_epilo, peak_time_epilo, fig_epilo, bg_mean_epilo = event_onset_epilo.analyse(viewing='3', bg_start=0, bg_length=5, channels=[4,5], resample_period=None, yscale='log', cusum_window=30)
```

Note the intensity spikes in the declining phase of the event that trigger the peak flux. Let's remove those in the next step.
